```python
import datetime
import pandas as pd
import pandas_datareader.data
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
dis = pandas_datareader.data.get_data_yahoo('DIS',
                                           start = datetime.datetime(2006,1,1))
```

```python
#get the Close timeseries
ts = dis['Close']
```

```python
ts[-10:]
```

```python
%matplotlib inline
```

```python
adj_close = dis['Adj Close']
```

```python
mavg = adj_close.rolling(window = 40, center = False).mean()
```

```python
ts.plot(label = 'DIS')
mavg.plot(label = 'mavg')
plt.legend()
```

```python
returns = adj_close / adj_close.shift(1) -1
pct_change = adj_close.pct_change()[1:]
```

```python
plt.hist(returns[1:],bins=15);
```

```python
#get mean and standard deviation
import numpy as np
np.mean(returns)
```
